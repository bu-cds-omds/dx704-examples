# Chart: An Example of Brownian Motion


```python
import matplotlib.pyplot as plt
import numpy as np

ts = range(0, 21)
for i in range(5):
    xs = [0]
    while(len(xs) < len(ts)):
        xs.append(xs[-1] + np.random.normal(0, 1))

    plt.plot(ts, xs, label=f"Sample {i+1}", marker="o")

plt.title("An Example of Brownian Motion")
plt.xlabel("Time")
plt.ylabel("Position")
plt.legend()
plt.show();
```
