# Chart: Real Gross Domestic Product (GDP)


```python
import matplotlib.pyplot as plt
import json
import pandas as pd

# data source: https://fred.stlouisfed.org/series/GDPC1


data_url = "https://fred.stlouisfed.org/graph/fredgraph.csv?bgcolor=%23ebf3fb&chart_type=line&drp=0&fo=open%20sans&graph_bgcolor=%23ffffff&height=450&mode=fred&recession_bars=on&txtcolor=%23444444&ts=12&tts=12&width=1071&nt=0&thu=0&trc=0&show_legend=yes&show_axis_titles=yes&show_tooltip=yes&id=GDPC1&scale=left&cosd=1947-01-01&coed=2025-01-01&line_color=%230073e6&link_values=false&line_style=solid&mark_type=none&mw=3&lw=3&ost=-99999&oet=99999&mma=0&fml=a&fq=Quarterly&fam=avg&fgst=lin&fgsnd=2020-02-01&line_index=1&transformation=lin&vintage_date=2025-06-04&revision_date=2025-06-04&nd=1947-01-01"
df = pd.read_csv(data_url)
df = df.dropna()
df = df.set_index("observation_date")
df = df.rename(columns={df.columns[0]: "price"})

df.plot()
plt.title("Real Gross Domestic Product")
plt.xlabel("Date")
plt.ylabel("Price")
plt.ylim(0)
plt.show();
```
