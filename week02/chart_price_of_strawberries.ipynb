# Chart: Price of Strawberries


```python
import matplotlib.pyplot as plt
import json
import pandas as pd

# data source: https://fred.stlouisfed.org/series/APU0000711415

data_url = "https://fred.stlouisfed.org/graph/fredgraph.csv?mode=fred&id=APU0000711415&cosd=1980-04-01&coed=2025-04-01&vintage_date=2025-06-04&revision_date=2025-06-04&nd=1980-04-01"

df = pd.read_csv(data_url)
df = df.dropna()
df = df.set_index("observation_date")
df = df.rename(columns={df.columns[0]: "price"})

df.plot()
plt.title("Average Price: Strawberries, Dry Pint in U.S. City Average")
plt.xlabel("Date")
plt.ylabel("Price")
plt.ylim(0)
plt.show();
```
