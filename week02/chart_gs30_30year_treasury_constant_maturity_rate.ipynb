# Chart: GS30 - 30-Year Treasury Constant Maturity Rate


```python
%pip install -q fredapi

import matplotlib.pyplot as plt
from fredapi import Fred
from google.colab import userdata

fred = Fred(api_key=userdata.get('FRED_API_KEY'))

data = fred.get_series('GS30')
data.plot()
plt.title("30-Year Treasury Constant Maturity Rate")
plt.xlabel("Date")
plt.ylabel("Price")
plt.ylim(0)
plt.show();
```
