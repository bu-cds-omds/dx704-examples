# Chart: European Option Values based on Price at Expiration


```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(0, 100, 100)
call_values = np.maximum(xs - 50, 0)
put_values = np.maximum(50 - xs, 0)

plt.plot(xs, call_values, label="Call Option")
plt.plot(xs, put_values, label="Put Option")

plt.legend()
plt.title("European Option Values Based on Price at Expiration")
plt.xlabel("Price")
plt.ylabel("Option Value")
plt.ylim(0)

plt.show();
```
