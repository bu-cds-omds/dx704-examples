# Chart: Position Distributions under Brownian Motion


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

xs = np.linspace(-10, 10, 100)

plt.figure(figsize=(12, 3))
for i in (1,2,3):
    plt.subplot(1, 3, i)

    plt.plot(xs, norm.pdf(xs, 0, i))
    plt.title(f"Probability Density Function of $X_{i}$")
    plt.ylim(0, 0.5)

plt.subplots_adjust(wspace=0.5)
plt.show();
```
