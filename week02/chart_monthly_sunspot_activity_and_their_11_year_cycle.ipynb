# Chart: Monthly Sunspot Activity and their 11 Year Cycle


```python
# Source: WDC-SILSO, Royal Observatory of Belgium, Brussels, DOI: https://doi.org/10.24414/qnza-ac80”.

import pandas as pd

df = pd.read_csv("https://www.sidc.be/SILSO/INFO/snmtotcsv.php",
                 names=["year", "month", "date_frac", "sunspots_mean", "sunspots_stdev", "observations", "status"],
                 sep=";")
df["date"] = pd.to_datetime(df["year"].astype(str) + "-" + df["month"].astype(str), format="%Y-%m")
df = df.set_index("date")
df["sunspots_mean"].plot();
```
