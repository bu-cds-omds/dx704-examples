```python
import google.genai as genai
from google.colab import userdata
```

```python
client = genai.Client(api_key=userdata.get('GEMINI_API_KEY'))
```

```python
# Rate limit documentation https://ai.google.dev/gemini-api/docs/rate-limits
model_name = 'gemini-2.0-flash'
```

```python
def get_response(contents):
    response = client.models.generate_content(model=model_name,
                                              contents=contents)
    return response.text
```

```python
get_response("Hi!")
```

```python
get_response("Complete the phrase: My friend is named Dip")
```

```python
get_response("Sir, do you bite thy thumb at me?")
```

```python
get_response("What is Jeffrey Considine's favorite number?")
```

```python
print(get_response("what does early snow mean in literature?"))
```

```python
print(get_response("Which food should I recommend for my dinner party? Ice cream because it is good and everyone likes it. Chicken because it always hits the sweet spot. Pizza because everyone likes it. No spinach I hate it. No liver because it stinks. So what is something new that should we make for dinner?"))
```
