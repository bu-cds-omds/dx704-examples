# Chart: Diminishing Utility from Variation


```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(0.001, 1, 1000)

plt.figure(figsize=(6, 6))
plt.plot(xs, xs**0.5, label="utility")

xs2 = np.array([0.1, 0.5])
ys2 = xs2**0.5
plt.plot(xs2, ys2, label="blended utility", marker="*")

plt.legend()
plt.title("Diminishing Utility from Variation")
plt.xlabel("Total Rewards")
plt.xlim(0)
plt.ylabel("Utility")
plt.ylim(0)

plt.show();
```
