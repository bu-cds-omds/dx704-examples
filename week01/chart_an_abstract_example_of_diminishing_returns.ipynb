# Chart: An Abstract Example of Diminishing Returns


```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(0.001, 1, 1000)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(xs, xs**0.5, label="utility")
plt.legend()
plt.title("Diminishing Utility")
plt.xlabel("Total Rewards")
plt.xlim(0)
plt.ylabel("Utility")
plt.ylim(0)

plt.subplot(1, 2, 2)
plt.plot(xs, 0.5 * xs**-0.5, label="marginal utility")
plt.legend()
plt.title("Diminishing Marginal Utility")
plt.xlabel("Total Rewards")
plt.xlim(0)
plt.ylabel("Utility")
plt.ylim(0, 10)

plt.show();
```
