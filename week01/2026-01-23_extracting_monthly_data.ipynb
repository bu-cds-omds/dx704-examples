```python
import matplotlib.pyplot as plt
import yfinance as yf
```

```python

data = yf.download("AAPL", start="2020-01-01", end="2025-12-31", progress=False)
data["Volume"].plot()
plt.ylim(0)
plt.show();
```

```python
close_data = data["Close"]
```

The `.last()` method returns rows with the last day of the month.

```python
close_data.resample("ME").last()
```

If you look at the last 5 days, the row labeled 2025-12-31 above has the close price of 2025-12-30 data.

```python
close_data[-5:]
```
