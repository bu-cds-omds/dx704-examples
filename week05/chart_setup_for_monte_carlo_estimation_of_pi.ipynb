```python
# https://chatgpt.com/share/6865e40d-669c-800c-be02-1890519e7c9c

import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create figure and axis
fig, ax = plt.subplots()

# Draw the 2x2 square
square = patches.Rectangle((-1, -1), 2, 2, linewidth=1, edgecolor='black', facecolor='none')
ax.add_patch(square)

# Draw the unit circle
circle = patches.Circle((0, 0), 1, linewidth=1, edgecolor='blue', facecolor='none')
ax.add_patch(circle)

# Set the limits and aspect
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect('equal')
ax.axhline(0, color='gray', linewidth=0.5)
ax.axvline(0, color='gray', linewidth=0.5)
ax.set_title('Unit Circle Inside a 2x2 Square')

plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```
